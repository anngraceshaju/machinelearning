```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

sns.set_theme(style="darkgrid")
```

```python
#load the simple dataset
df=pd.read_csv("pass.csv")
print("Dataset:")
print(df.head())

#visualise the data points
plt.figure(figsize=(8,5))
plt.scatter(df['hours'], df['pass'], c=df['pass'], cmap='bwr', edgecolor='k')
plt.xlabel('hours studied')
plt.ylabel('pass(1)/Fail(0)')
plt.title('hours vs Result')
plt.show()
```

```python
#prepare features and target
x = df[['hours']]  #input
y = df[['pass']]  #output
model=LogisticRegression()
model.fit(x,y)
print("model trained successfully.")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['hours'],df['pass'],c=df["pass"],cmap="bwr",label='Data points')

#generate a range of values for the curve
x_test_range=np.linspace(0,10,100). reshape(-1,1)
y_prob_range=model.predict_proba(x_test_range)[:,1]

plt.plot(x_test_range,y_prob_range,color="green",linewidth=3,label='Logistic curve')
plt.axhline(y=0.5,color="gray",linestyle='--',label='Decision Boundary(0.5)')
plt.xlabel('hours studied')
plt.ylabel('probability of passing')
plt.title('Logistic Regression sigmoid curve')
plt.legend()
plt.show()
```

```python
try:
    #get user input
    user_hours=float(input("enter study hours (0,10):"))

    #prepare input for model
    x_user = np.array([[user_hours]])

    #make prediction
    prediction = model.predict(x_user)[0]
    probability = model.predict_proba(x_user)[0][1]

    #ouput result
    if prediction ==1:
        print(f"\nResult: pass")
        print(f"confidence: {(probability)*100:.2f}%")  
    else:
        print(f"\nResult: fail")
        print(f"confidence: {(1-probability)*100:.2f}%")
except ValueError:
    print("invaild input. please enter a numerical value for hours:.")
```
